```python
# 1900년 이전 생일의 화자 생년 바꾸는 함수
def birthyear_1052(values):
    if values<=1100:
        return values +900
    elif values<1700:
        return values+300
    else:
        return values
# 함수 적용해 1900년 이상으로 모두 변경
main_df['birth_year']=main_df['birth_year'].apply(birthyear_1052)
```
