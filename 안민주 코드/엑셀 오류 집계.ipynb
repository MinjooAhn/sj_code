```python
import pandas as pd
import re
import numpy as np
```

```python
# 영어 4개 파일 통합 통계
df1 = pd.read_excel(r"C:\Users\paula\Desktop\Sejong Data project\외국인\오류 퍼센트_태국어제외\영어\영어 원본 통합\ENRB 1-150.xlsx")
df2 = pd.read_excel(r"C:\Users\paula\Desktop\Sejong Data project\외국인\오류 퍼센트_태국어제외\영어\영어 원본 통합\ENRC 1-150.xlsx")
df3 = pd.read_excel(r"C:\Users\paula\Desktop\Sejong Data project\외국인\오류 퍼센트_태국어제외\영어\영어 원본 통합\ENQC 226-300.xlsx")
df4 = pd.read_excel(r"C:\Users\paula\Desktop\Sejong Data project\외국인\오류 퍼센트_태국어제외\영어\영어 원본 통합\ENQB 226-300.xlsx")
# 오류 종류
error_types = ['O1', 'O2', 'O3', 'O4', 'O5', 'O6',
               'Oi', 'C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'C8', 'Ci',
               'N1', 'N2', 'R1', 'R2', 'R3', 'R4', 'R5', 'R6', 'R7', 'R8', 'R#m',
               'S', 'F', 'P']

# 오류가 적힌 컬럼 인덱싱 
error_column1 = df1.columns[7:26]
error_column2 = df1.columns[7:31]
error_column3 = df1.columns[7:36]
error_column4 = df1.columns[7:36]

# 오류 종류 데이터프레임 제작
error_counts = pd.DataFrame({'Error_Type': error_types})

# 오류 빈도 추출
for index, row in error_counts.iterrows():
    error_type = row['Error_Type']
    pattern = rf'(?i)\b{error_type}\b'
    # 4개 파일 오류 합하기
    error_counts.at[index, 'Frequency'] = (
        df1[error_column1].astype(str).apply(lambda x: x.str.count(pattern)).sum().sum() +
        df2[error_column2].astype(str).apply(lambda x: x.str.count(pattern)).sum().sum() +
        df3[error_column3].astype(str).apply(lambda x: x.str.count(pattern)).sum().sum() +
        df4[error_column4].astype(str).apply(lambda x: x.str.count(pattern)).sum().sum()
    )

# 빈도 내림차순으로 정렬
combined_df = error_counts.sort_values(by="Frequency", ascending=False)
combined_df['Percentage'] = (combined_df["Frequency"] / combined_df["Frequency"].sum()) * 100
# 정렬된 데이터프레임을 엑셀로 출력
combined_df.to_excel(
    r"C:\Users\paula\Desktop\Sejong Data project\외국인\영어 분포통계.final.xlsx",
    index=False,
    engine='xlsxwriter'
)
```
