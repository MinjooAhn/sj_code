```python
import pandas as pd
import json

excel_file_path = ""
data_frame = pd.read_excel(excel_file_path)

json_data = []
current_level = None

for _, row_data_frame in data_frame.iterrows():
    if pd.isna(row_data_frame["LevelNo"]):
        if json_data:
            json_file_path = "".format(row_data["FileName"])
            with open(json_file_path, 'w', encoding="utf-8") as json_file:
                json.dump(json_data, json_file, indent=4, ensure_ascii=False)
            print(f"Data successfully converted and saved to '{json_file_path}'")
            json_data = []
            current_level = None
    else:
        if current_level is None:
            current_level = row_data_frame["LevelNo"]
        json_template = {
        "FileName": "",
        "LevelNo":"",
        "LessonNo":None,
        "PageNo":None,
        "Type":"",
        "SetNo":None,
        "Domain": {
                "Category":"",
                "Item":"",
        },
        "Sentence": {
            "SentenceID":"",
            "SpeakerID":None,
            "SpeakerGender":"",
            "SentenceNo":"",
            "Text":"",
            "Eng":"",
            "Intent": {
                "Category":"",
                "Item":"",
            },
            "Entity":""
        }
    }
        if pd.notna(row_data_frame["LevelNo"]): 
            row_data = json_template.copy()

            row_data["LevelNo"] = row_data_frame["LevelNo"]
            
            if pd.notna(row_data_frame["LessonNo"]):
                row_data["LessonNo"] = int(row_data_frame["LessonNo"])

            if pd.notna(row_data_frame["PageNo"]):
                row_data["PageNo"] = int(row_data_frame["PageNo"])

            row_data["Type"] = row_data_frame["Type"]

            if pd.notna(row_data_frame["SetNo"]):
                row_data["SetNo"] = int(row_data_frame["SetNo"])
            if pd.notna(row_data_frame["SpeakerID"]):
                row_data["Sentence"]["SpeakerID"] = int(row_data_frame["SpeakerID"])
            row_data["Sentence"]["SpeakerGender"] = row_data_frame["SpeakerGender"]
            row_data["Sentence"]["SentenceNo"] = row_data_frame["SentenceNo"]
            row_data["Sentence"]["Text"] = row_data_frame["Text"]
            row_data["Sentence"]["Eng"] = row_data_frame["Eng"]
            row_data["Domain"]["Category"] = row_data_frame["DomainCategory"]
            row_data["Domain"]["Item"] = row_data_frame["DomainItem"]
            row_data["Sentence"]["Intent"]["Category"] = row_data_frame["IntentCategory"]
            row_data["Sentence"]["Intent"]["Item"] = row_data_frame["IntentItem"]
            row_data["Sentence"]["Entity"] = row_data_frame["Entity1"]
            row_data["Sentence"]["SentenceID"] = f"{row_data['LevelNo']}_{row_data['LessonNo']}_{row_data['PageNo']}_{row_data['Type']}_{row_data['SetNo']}_{row_data['Sentence']['SpeakerID']}_{row_data['Sentence']['SpeakerGender']}_{row_data['Sentence']['SentenceNo']}"
            row_data["FileName"] = f"{row_data['LevelNo']}_{row_data['LessonNo']}_{row_data['PageNo']}_{row_data['Type']}_{row_data['SetNo']}_{row_data['Domain']['Category']}"
            json_data.append(row_data)


if json_data:
    json_file_path = "".format(row_data["FileName"])
    with open(json_file_path, 'w', encoding="utf-8") as json_file:
        json.dump(json_data, json_file, indent=4, ensure_ascii=False)
    print(f"Data successfully converted and saved to '{json_file_path}'")
```
