```python
# Errors in error types
import pandas as pd

data = {
    'File_Name': [],
    '오류': [],
    '철자전사': [],
    '발음전사': []
}

df2 = pd.DataFrame(data)

excel_file_path = r""
sheet_name = ''  

# List of error types to find in the Excel file
error_types = ['O1', 'C4', 'O2', 'C5', 'N1', 'C1']

# Read the Excel file
df = pd.read_excel(excel_file_path, sheet_name=sheet_name, engine='openpyxl')

# Create a writer for the output Excel file
writer = pd.ExcelWriter(r"")

# Loop through each error type
for error_type_to_find in error_types:
    # Create a new DataFrame for each error type
    df_temp = pd.DataFrame(data)

    for row_idx, row in df.iterrows():
        for col in df.columns[8:28]:
            cell_value = str(row[col])
            if error_type_to_find in cell_value:
                # Log the filename from the second column
                File_name = str(row[df.columns[1]])

                # Log the word which is 1 cell above the cell with the error type
                Word = str(df.iloc[row_idx - 2, df.columns.get_loc(col)])

                # Log the misspelling which is 2 cells above the cell with the error type
                Misspelling = str(df.iloc[row_idx - 1, df.columns.get_loc(col)])

                # Append the data to the temporary DataFrame
                df_temp.loc[len(df_temp)] = [File_name, error_type_to_find, Word, Misspelling]

    print(f"Error Type: {error_type_to_find}")
    print(df_temp)
    df_temp.to_excel(writer, sheet_name=error_type_to_find, index=False)

# Save and close the Excel writer after all sheets are written
writer.close()
```

```python
# Error Frequency
import pandas as pd

excel_file_path = r""
sheet_name = '' 

# List of error types to find in the Excel file
error_types = ['O1', 'O2', 'O3', 'O4', 'O5', 'O6',
                'Oi', 'C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'C8', 'Ci',
                'N1', 'N2', 'R1', 'R2', 'R3', 'R4', 'R5', 'R6', 'R7', 'R8', 'R#m',
                'S', 'F', 'P'
]

# Read the Excel file
df = pd.read_excel(excel_file_path, sheet_name=sheet_name, engine='openpyxl')

# Create a writer for the output Excel file
writer = pd.ExcelWriter(r"")

# Initialize a dictionary to store the frequency count for each error type
frequency_count = {error_type: 0 for error_type in error_types}

# Loop through each error type
for error_type_to_find in error_types:
    for row_idx, row in df.iterrows():
        for col in df.columns[8:28]:
            cell_value = str(row[col])
            if error_type_to_find in cell_value:
                # Increment the frequency count for the current error type
                frequency_count[error_type_to_find] += 1

# Calculate the total frequency
total_frequency = sum(frequency_count.values())

# Calculate the percentage for each error type
percentage_count = {error_type: (count / total_frequency) * 100 for error_type, count in frequency_count.items()}

df_result = pd.DataFrame({
    'error_type': list(frequency_count.keys()),
    'Frequency': list(frequency_count.values()),
    'Percentage': list(percentage_count.values())
})

df_result = df_result.sort_values(by='Frequency', ascending=False)
print("Result:")
print(df_result)
df_result.to_excel(writer, sheet_name='Result', index=False)
writer.close()
```

```python
# Errors containing at least 1 from error type
import pandas as pd

data = {
    'File_Name': [],
    '5대 오류': [],
    '총 오류': [],
    '철자전사': [],
    '발음전사': []
}

excel_file_path = r""
sheet_name = ''  

# List of error types to find in the Excel file
error_types = ['O1', 'C4', 'O2', 'C5', 'C1']

# Read the Excel file
df = pd.read_excel(excel_file_path, sheet_name=sheet_name, engine='openpyxl')

# Create a writer for the output Excel file
writer = pd.ExcelWriter(r"")

# Loop through each error type
for error_type_to_find in error_types:
    df_temp = pd.DataFrame(data)

    for row_idx, row in df.iterrows():
        for col in df.columns[8:28]:
            cell_value = str(row[col])
            errors_in_cell = cell_value.split()
            all_errors_in_cell=[]

            for error in errors_in_cell:
                all_errors_in_cell.append(error)

            if error_type_to_find in all_errors_in_cell and len(all_errors_in_cell) >1 :
                # Log the filename from the second column
                File_name = str(row[df.columns[1]])

                # Log the word which is 1 cell above the cell with the error type
                Word = str(df.iloc[row_idx - 2, df.columns.get_loc(col)])

                # Log the misspelling which is 2 cells above the cell with the error type
                Misspelling = str(df.iloc[row_idx - 1, df.columns.get_loc(col)])

                # Append the data to the temporary DataFrame
                df_temp.loc[len(df_temp)] = [File_name, error_type_to_find,', '.join(all_errors_in_cell), Word, Misspelling]

    print(f"Error Type: {error_type_to_find}")
    print(df_temp)

    df_temp.to_excel(writer, sheet_name=error_type_to_find, index=False)

# Save and close the Excel writer after all sheets are written
writer.close()
```
